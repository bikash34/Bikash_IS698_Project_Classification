```python
#import numpy, pandas and sklearn train_test_split, DecisionTreeClassifier & accuracy_score modules

import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn import tree
```

```python
#import data into pandas dataframe (in this case dataframe is schoolFires_data) using Pandas read_csv() method

path_to_file = 'C:/Users/bikash/Desktop/Classification_Dataset_Integrated_2010_2014.csv'
schoolFires_data = pd.read_csv(path_to_file,sep=',',header=0)
```

```python
#check the length (number of rows) of dataframe:schoolFires_data  

print ("Dataset Length:: ", len(schoolFires_data))
```

```python
#check the length & dimensions (number of columns) of dataframe:schoolFires_data

print ("Dataset Shape:: ", schoolFires_data.shape) 
```

```python
# print top 5 rows of dataframe:schoolFires_data

print("Dataset:: ")
schoolFires_data.head()
```

```python
#dataset slicing (divide data into features set & target set)

features = schoolFires_data[['Median_Income', 'UnemployedYouth_Population']].values

target = schoolFires_data['SchoolFire_Class'].values
```

```python
# split data into training set and test set in the ratio of 70:30

features_train, features_test, target_train, target_test = train_test_split( features, target, test_size = 0.3)
```

```python
# build Decision Tree Classifier with criterion gini index without specifying max_depth

clf = tree.DecisionTreeClassifier()
clf = clf.fit(features_train, target_train)
```

```python
# visulize the tree using graphviz

import graphviz 
dot_data = tree.export_graphviz(clf, out_file=None) 
graph = graphviz.Source(dot_data) 
graph.render("SchoolFiresData") 


dot_data = tree.export_graphviz(clf, out_file=None, 
                         feature_names=['Median_Income', 'UnemployedYouth_Population'],  
                         class_names= schoolFires_data['SchoolFire_Class'].values,  
                         filled=True, rounded=True,  
                         special_characters=True)  
graph = graphviz.Source(dot_data)  
graph 
```

```python
# predict classes for test set and print the predicted classes

target_predict = clf.predict(features_test)
target_predict
```

```python
# predict class for the following randomly selected test set
  
clf.predict([[201793, 595]])		#(for which class = Low_Frequency)
```

```python
# calculate accuracy score for Decision Tree classifier with criterion as gini index

print ("Accuracy is ", accuracy_score(target_test,target_predict)*100)
```
